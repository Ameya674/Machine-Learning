```python
import pandas as pd
df = pd.read_csv('dt.csv')
df.head(2)
```

```python
i = df.drop('salary(100)' , axis = 'columns')
i
```

```python
test = df['salary(100)']
test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
c = LabelEncoder()
j = LabelEncoder()
d = LabelEncoder()
```

```python
i['c'] = c.fit_transform(i['company'])
i['j'] = j.fit_transform(i['job'])
i['d'] = d.fit_transform(i['degree'])
```

```python
i 
```

```python
train = i.drop(['company','job','degree'] , axis = 'columns')
train
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(train,test)
```

```python
model.score(train,test)
```

```python
model.predict([[2,2,1]])
```
