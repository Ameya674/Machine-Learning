```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('km.csv')
df.head(3)
```

```python
plt.scatter(df['age'] , df['income'])
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
yp = km.fit_predict(df[['age','income']] , )
yp
```

```python
df['cluster'] = yp
df
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
```

```python
plt.scatter(df0.age , df0.income)
plt.scatter(df1.age , df1.income)
plt.scatter(df2.age , df2.income)
```

```python
mmx = MinMaxScaler()

mmx.fit(df[['income']])
df['income'] = mmx.transform(df[['income']])

mmx.fit(df[['age']])
df['age'] = mmx.transform(df[['age']])

df
```

```python
km = KMeans(n_clusters=3)
yp = km.fit_predict(df[['age','income']])
yp
```

```python
df['cluster'] = yp
df
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
```

```python
plt.scatter(df0['age'] , df0['income'])
plt.scatter(df1['age'] , df1['income'])
plt.scatter(df2['age'] , df2['income'])
```
